```julia
function AR1(ρ = .95; dist=Normal(0,0.5), N=10000)
    ϵ = rand(dist,N) .-mean(dist)
    β = 0
#     Y = [0.0 for i = 1:N]
    Y=zeros(N)
    # Y[1]=0
    irf= zeros(N)
    irf[1]=1
    for t in 2:(N)
        Y[t] = β + ρ*Y[t-1] + ϵ[t]
        irf[t] = ρ*irf[t-1]
     end
    return(Y,ϵ,irf)
end

function AR1Plot(ρ=.95;N=10000,n=500,dist=Normal(0,0.5))
    title = nameof(typeof(dist))
    (Y,ϵ) = AR1(ρ)
    T = N-n+1
    plot(repeat((1:n)',2),[zeros(1,n);ϵ[T:N]'],label="",color=:grey, alpha=.5)
    plot!(1:n,ϵ[T:N],markershape=:circle,linewidth=2,label="",color=:grey, alpha=.5)
    plot!(1:n,Y[T:N],markershape=:circle,linewidth=2,label="",color=:blue, alpha=.6)
    
    return plot!(title="AR(1) with ρ = $ρ & $title")
end

function AR1PlotIRF(ρ=.95;n=100)
    (Y,ϵ,irf) = AR1(ρ,N=n)
    
    plot(1:n,irf[1:n],label="",linewidth=1.5)
    plot!(1:1,1:1,color=:grey, markershape=:circle, alpha=.5,label="",linewidth=1.5)
    return plot!(title="IRF with ρ = $ρ")
end
```

```julia
# N=10000
# n=500
ρ =.95
dist=Normal(0,0.5)
AR1Plot(ρ)
```

```julia
#2.3

function AR2(ρ1 = .95,ρ2=.95; dist=Normal(0,1), N=10000)
    ϵ = rand(dist,N) .-mean(dist)
    β = 0
#     Y = [0.0 for i = 1:N]
    Y=zeros(N)
    # Y[1]=0
    irf= zeros(N)
    irf[1]=1
    for t in 3:(N)
        Y[t] = β + ρ1*Y[t-1] +ρ2*Y[t-2] + ϵ[t]
        irf[t] = ρ1*irf[t-1] + ρ2*irf[t-2]
     end
    return(Y,ϵ,irf)
end

function AR2Plot(ρ1=.95,ρ2=.95;N=10000,n=500,dist=Normal(0,0.5))
    title = nameof(typeof(dist))
    (Y,ϵ) = AR2(ρ1,ρ2)
    T = N-n+1
    plot(repeat((1:n)',2),[zeros(1,n);ϵ[T:N]'],label="",color=:grey, alpha=.5)
    plot!(1:n,ϵ[T:N],markershape=:circle,linewidth=2,label="",color=:grey, alpha=.5)
    plot!(1:n,Y[T:N],markershape=:circle,linewidth=2,label="",color=:blue, alpha=.6)
    
    return plot!(title="AR(1) with ρ1 = $ρ1, ρ2 = $ρ2 $title")
end

function AR2PlotIRF(ρ1=.95,ρ2=.95;n=100)
    (Y,ϵ,irf) = AR2(ρ1,ρ2,N=n)
    
    plot(1:n,irf[1:n],label="",linewidth=1.5)
    plot!(1:1,1:1,color=:grey, markershape=:circle, alpha=.5,label="",linewidth=1.5)
    return plot!(title="IRF with ρ1 = $ρ1, ρ2 = $ρ2")
end
```

```julia
p11=AR2Plot(.5,.5)
p12=AR2Plot(.5,.9)
p13=AR2Plot(.5,-.5)
p14=AR2Plot(.5,-.9)
p21=AR2Plot(.9,.5)
p22=AR2Plot(.9,.9)
p23=AR2Plot(.9,-.5)
p24=AR2Plot(.9,-.9)
plot(p11,p12,p13,p14,p21,p22,p23,p24,layout=(2,4))
# plot(p11)
```

```julia
function OutputDev(ρA = 0,α = 0.66;dist=Normal(0,0.5), N=10000)
#     ϵ = rand(dist,N) .-mean(dist)
    ϵ = zeros(N)
    ϵ[3] = 1.5
    Ỹ=zeros(N)
    # Y[1]=0
    irf= zeros(N)
    irf[1]=1
    for t in 3:(N)
        Ỹ[t] = (α+ρA)*Ỹ[t-1] -α*ρA*Ỹ[t-2] + (1-α)*ϵ[t]
        irf[t] = (α+ρA)*irf[t-1] -α*ρA*irf[t-2]
     end
    return(Ỹ,ϵ,irf)
end

function OutputDevPlot(ρA=0,α = 0.66;N=10000,n=100,dist=Normal(0,0.5))
    title = nameof(typeof(dist))
    (Ỹ,ϵ) = OutputDev(ρA,α)
    T = N-n+1
#     plot(repeat((1:n)',2),[zeros(1,n);ϵ[T:N]'],label="",color=:grey, alpha=.5)
#     plot(1:n,ϵ[1:n],markershape=:circle,linewidth=1,label="",color=:grey, alpha=.1)
    plot(1:n,Ỹ[1:n],markershape=:circle,linewidth=1,label="",color=:blue, alpha=.5,yticks = 0:0.1:1.5,markersize=2.5)
    
    return plot!(title="Output Deviation AR(2) with ρA = $ρA & α = $α")
end

function OutputDevPlotIRF(ρA=0;n=100)
    (Ỹ,ϵ,irf) = OutputDev(ρA,N=n)
    
    plot(1:n,irf[1:n],label="",linewidth=1.5)
    plot!(1:1,1:1,color=:grey, markershape=:circle, alpha=.5,label="",linewidth=1.5,yticks = 0:0.1:1.5)
    return plot!(title="IRF with ρA = $ρA")
end

```
